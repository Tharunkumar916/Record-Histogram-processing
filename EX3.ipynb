```python
# Import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Read the image in greyscale form

img = cv2.imread('parrot.jpg', cv2.IMREAD_COLOR)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
#Display the Image in  Greyscale form
img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)
plt.imshow(img_gray,cmap='gray')
plt.show()
```

```python
plt.hist(img.ravel(),256,range = [0,256])
plt.show()
```

```python
img_eq=cv2.equalizeHist(img_gray)
```

```python

plt.hist(img_eq.ravel(),256,range=[0,256])
plt.show()
```

```python
plt.imshow(img_eq,cmap='gray')
plt.show()
```

```python
img = cv2.imread('parrot.jpg',cv2.IMREAD_COLOR)
```

```python
img_hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
```

```python
img_hsv[:,:,2]=cv2.equalizeHist(img_hsv[:,:,2])
```

```python
img_e=cv2.cvtColor(img_hsv,cv2.COLOR_HSV2BGR)
```

```python
plt.imshow(img_e[:,:,::-1]);plt.title('Equalized Image');plt.show()
```

```python
plt.hist(img_e.ravel(),256,range=[0,256]);plt.title('Histogram Equalized');plt.show()
```

```python
#Disply the images.
#plt.figure(figsize=(20,10))
plt.subplot(221); plt.imshow(img[:,:,::-1]);plt.title('Original Color Image')
plt.subplot(222); plt.imshow(img_e[:,:,::-1]); plt.title('Equalized Image')
plt.subplot(223); plt.hist(img.ravel(),256,range= [0,256]);plt.title('Original Image')
plt.subplot(224); plt.hist(img_e.ravel(),256,range = [0,256]); plt.title('Histogram Equalized'); plt.show()
```

```python
#Display the histograms.
plt.figure(figsize=[15,4])
plt.subplot(121); plt.hist(img.ravel(),256,range =[0,256]); plt.title('Original Image')
plt.subplot(122); plt.hist(img_e.ravel(),256,range=[0,256]); plt.title('Histogram Equalized')
```
